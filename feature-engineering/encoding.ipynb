### Nominal/OHE Encoding
One hot encoding, also known as nominal encoding, is a technique used to represent categorical data as numerical data, which is more suitable for machine learning algorithms. In this technique, each category is represented as a binary vector where each bit corresponds to a unique category. For example, if we have a categorical variable "color" with three possible values (red, green, blue), we can represent it using one hot encoding as follows:

1. Red: [1, 0, 0]
2. Green: [0, 1, 0]
3. Blue: [0, 0, 1]

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
## cCreate a single dataframe
df = pd.DataFrame({
    'color' : ['red','blue','green','green','red','blue' ]
})
```

```python
df.head()
```

```python
## Create an instamce one hot encoder
encoder=OneHotEncoder()
```

```python
## perforn fit and then transform
encoded=encoder.fit_transform(df[['color']]).toarray()
```

```python
import pandas as pd
encoder_df = pd.DataFrame(encoded,columns=encoder.get_feature_names_out())

```

```python
encoder_df
```

```python
## for any new data
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df,encoder_df],axis=1)
```

#### Ordinal Encoding


It is used to encode categorical data that have instrisic order or ranking. In this technique each category is assigned a nujmerical value based on its positiobn in the order . For example, if we have a categorical variable "educational level" with the possible values (high school, collage, graduate, post-graduate) we can represent using ordinal encoding as follow:

1. High school:1
2. College:2
3. Gradute:3
4. Post-Graduate:4

```python
## Ordinal Encoding
from sklearn.preprocessing import OrdinalEncoder
import pandas as pd
```

```python
df3 = pd.DataFrame({
    'size':['small','medium','large','medium','small','large']
})
```

```python
df3
```

```python
## create an instance of Ordinal Encoder and then fit_transform
encoder = OrdinalEncoder(categories=[['small','medium','large']])
```

```python
encoder.fit_transform(df3[['size']])
```

```python
encoder.transform([['small']])
```

### Target Guided Ordinal Encoding

It is a technique used to encode ctaegorical variable based on thier relationship with the target varibale. This encoding technique is useful when we have categorical variable with larger number of unique categories , and we want to use this variable as a feature in our machine learning model.

in Target Guided Ordinal Encoding , we replace each category in the categorical with anumerical value based on the mean or median of thr target variable for that category. This creates a monotonic relationship between the ctaegorical variable and the target variable, which can improve the predective power of our model


```python
import pandas as pd
df = pd.DataFrame({
    'city':['New York','London','Paris','Tokyo','New York','Paris'],
    'price':[200,150,300,250,180,320]
})
```

```python
df
```

```python
mean_price=df.groupby('city')['price'].mean().to_dict()
```

```python
mean_price
```

```python
df['city_encoded'] = df['city'].map(mean_price)
```

```python
df[['city_encoded','price']]
```
